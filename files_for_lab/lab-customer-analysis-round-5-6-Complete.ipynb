### 2. Encoding (categorical)

```python
df_cat=df.select_dtypes("object")
df_cat.head()
```

```python
df_cat=df_cat.drop(["Effective To Date","Customer"], axis=1)
```

```python
df_cat.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first').fit(df_cat)
encoded = encoder.transform(df_cat).toarray()
encoded_names = list(encoder.get_feature_names())

df_cat_encoded = pd.DataFrame(encoded,columns=encoded_names)
```

```python
df_cat_encoded.head()
```

```python
x=pd.concat([df_cat_encoded,x_num_normalised],axis=1)
```

```python
x.head(100)
```

## Linear Regression

- Train-test split.
- Apply linear regression.

```python
#Train-test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
#Apply linear regression
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
model = regression.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
print(y_pred)
```

## Model Validation

- Description:
  - R2.
  - MSE.
  - RMSE.
  - MAE.

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

R2=r2_score(y_test, y_pred)
print("R2 is: ", R2)
MSE=mean_squared_error(y_test, y_pred)
print("MSE is: ", MSE)
RMSE=math.sqrt(MSE)
print("RMSE is: ", RMSE)
MAE=mean_absolute_error(y_test, y_pred)
print("MAE is: ", MAE)
```
